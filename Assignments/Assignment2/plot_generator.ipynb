```python
import matplotlib.pyplot as plt
```

```python
def plot_train_size_curves(train_size_data, train_acc_data1, train_acc_data2, xlabel, ylabel, label1, label2, title, location='upper left'):
    x_values = list(range(len(train_size_data)))
    plt.plot(x_values, train_acc_data1, label=label1)
    plt.plot(x_values, train_acc_data2, label=label2)
    plt.xticks(x_values, train_size_data)
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.legend(loc=location)
    plt.show()
    
```

```python
train_sizes = [0.05, 0.1, 0.25, 0.5, 1]
train_accuracy_60 = [43.11, 51.14, 61.58, 71.59, 84.21]
test_accuracy_60 = [42.81, 49.38 ,57.53, 63.28, 68.56]
train_accuracy_100 = [47.2, 53.5, 66.86, 78.93, 96.02]
test_accuracy_100 = [45.49, 50.97, 59.18, 64.73, 70.37]

plot_train_size_curves(train_sizes, train_accuracy_60, train_accuracy_100, 'Train Ratio Size','Train Accuracy', 'Train Accuracy for 60 epochs', 'Train Accuracy for 100 epochs', 'Train Accuracy for different training sizes and epochs')
plot_train_size_curves(train_sizes, test_accuracy_60, test_accuracy_100, 'Train Ratio Size', 'Test Accuracy', 'Test Accuracy for 60 epochs', 'Test Accuracy for 100 epochs', 'Test Accuracy for different training sizes and epochs')
```

```python
def plot_patch_data(patch_sizes, y_axis_data, xlabel, ylabel, labels_data, title, location = 'upper right'):
    x_values = list(range(len(patch_sizes)))
    for ydata, label_data in zip(y_axis_data, labels_data):
        plt.plot(x_values, ydata, label=label_data)
    plt.xticks(x_values, patch_sizes)
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.legend(loc=location)
    plt.show()    
```

```python
patch_sizes_data = [4, 8, 16]


train_acc_overlap_data_60 = [66.84, 66.88, 64.66]
test_acc_overlap_data_60 = [62.71, 60.75, 57.61]

train_acc_non_overlap_data_60 = [65.36, 62.89, 62.68 ]
test_acc_non_overlap_data_60 = [59.96, 57.43, 55.1]

train_acc_overlap_data_100 = [74,  73.34,  72.06]
test_acc_overlap_data_100 = [66.16, 63.95, 58.66]

train_acc_non_overlap_data_100 = [71.54, 69.01, 69.74]
test_acc_non_overlap_data_100 = [62.71, 59.38, 56.79]

# overlap vs non-overlap comparision within same epoch
plot_patch_data(patch_sizes_data, [train_acc_overlap_data_60, train_acc_non_overlap_data_60], 'Patch Sizes', 'Train Accuracy', ['Train accuracy for overlap patches', 'Train accuracy for non-overlap patches'], 'Train Accuracy (60 epochs) for different patch sizes and types.')
plot_patch_data(patch_sizes_data, [test_acc_overlap_data_60, test_acc_non_overlap_data_60], 'Patch Sizes', 'Test Accuracy', ['Test accuracy for overlap patches', 'Test accuracy for non-overlap patches'], 'Test Accuracy (60 epochs) for different patch sizes and types.')

plot_patch_data(patch_sizes_data, [train_acc_overlap_data_100, train_acc_non_overlap_data_100], 'Patch Sizes', 'Train Accuracy', ['Train accuracy for overlap patches', 'Train accuracy for non-overlap patches'], 'Train Accuracy (100 epochs) for different patch sizes and types.')
plot_patch_data(patch_sizes_data, [test_acc_overlap_data_100, test_acc_non_overlap_data_100], 'Patch Sizes', 'Test Accuracy', ['Test accuracy for overlap patches', 'Test accuracy for non-overlap patches'], 'Test Accuracy (100 epochs) for different patch sizes and types.')


# all 4 for train and test
plot_patch_data(patch_sizes_data, [train_acc_overlap_data_60, train_acc_non_overlap_data_60, train_acc_overlap_data_100, train_acc_non_overlap_data_100], 'Patch Sizes', 'Train Accuracy', ['Train accuracy (60 epochs) for overlap patches', 'Train accuracy (60 epochs) for non-overlap patches', 'Train accuracy (100 epochs) for overlap patches', 'Train accuracy (100 epochs) for non-overlap patches',], 'Train Accuracy for different epochs, patch sizes and types.')
plot_patch_data(patch_sizes_data, [test_acc_overlap_data_60, test_acc_non_overlap_data_60, test_acc_overlap_data_100, test_acc_non_overlap_data_100], 'Patch Sizes', 'Test Accuracy', ['Test accuracy (60 epochs) for overlap patches', 'Test accuracy (60 epochs) for non-overlap patches', 'Test accuracy (100 epochs) for overlap patches', 'Test accuracy (100 epochs) for non-overlap patches',], 'Test Accuracy for different epochs, patch sizes and types.')
```
